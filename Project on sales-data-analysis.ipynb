```python
# Import necessary libraries
import pandas as pd
```

```python
# Upload the Superstore Excel file
from google.colab import files
uploaded = files.upload()
```

```python
# Load the Excel file into a DataFrame
df = pd.read_excel(next(iter(uploaded)))
```

```python
# Display the first few rows
df.head()
```

```python
# Show general information about the dataset
df.info()  #This gives you an overview—number of entries, column names, non-null counts, and data types.
```

```python
# Check for missing/null values
df.isnull().sum()   #This tells you which columns have missing data so we know what needs cleaning.
```

```python
# Basic descriptive statistics for numerical columns
df.describe()    #This shows averages, minimums, maximums, and standard deviations for columns like Sales, Profit, Quantity, etc.
```

**Total Sales, Profit & Orders**

```python
#calculate basic KPIs (Key Performance Indicators):
# Total Sales
total_sales = df['Sales'].sum()
```

```python
# Total Profit
total_profit = df['Profit'].sum()
```

```python
# Total Orders
total_orders = df['Order ID'].nunique()
```

```python
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Total Orders: {total_orders}")
```

**Top 10 Products by Sales**

```python
# Group by product name and sum sales
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products
```

**Sales by Category and Sub-Category**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set plot style
sns.set(style="whitegrid")

```

```python
# Sales by Category
plt.figure(figsize=(8, 5))
sns.barplot(data=df.groupby('Category')['Sales'].sum().reset_index(), x='Category', y='Sales', palette='Blues_d')
plt.title("Total Sales by Category")
plt.ylabel("Sales ($)")
plt.xlabel("Category")
plt.show()
```

```python
# Sales by Sub-Category
plt.figure(figsize=(12, 6))
sns.barplot(data=df.groupby('Sub-Category')['Sales'].sum().reset_index().sort_values('Sales', ascending=False),
            x='Sub-Category', y='Sales', palette='coolwarm')
plt.title("Total Sales by Sub-Category")
plt.xticks(rotation=45)
plt.ylabel("Sales ($)")
plt.xlabel("Sub-Category")
plt.show()
```

**Monthly Sales Trend**

```python
# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Create a new column for Month-Year
df['Month-Year'] = df['Order Date'].dt.to_period('M')
```

```python
# Group by Month-Year and sum sales
monthly_sales = df.groupby('Month-Year')['Sales'].sum().reset_index()
```

```python
# Plot
plt.figure(figsize=(14, 6))
monthly_sales['Month-Year'] = monthly_sales['Month-Year'].astype(str)
sns.lineplot(data=monthly_sales, x='Month-Year', y='Sales', marker='o', color='teal')
plt.title("Monthly Sales Trend")
plt.xticks(rotation=45)
plt.ylabel("Sales ($)")
plt.xlabel("Month-Year")
plt.tight_layout()
plt.show()
```

**Sales by Region**

```python
# Group by Region
region_sales = df.groupby('Region')['Sales'].sum().reset_index().sort_values('Sales', ascending=False)
```

```python
# Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=region_sales, x='Region', y='Sales', palette='viridis')
plt.title("Sales by Region")
plt.ylabel("Sales ($)")
plt.xlabel("Region")
plt.show()
```

**Profit vs Sales by Category**

```python
# Group by Category and calculate both Sales and Profit
category_summary = df.groupby('Category')[['Sales', 'Profit']].sum().reset_index()
```

```python
# Plot
category_summary.plot(kind='bar', x='Category', figsize=(8, 5))
plt.title("Sales vs Profit by Category")
plt.ylabel("Amount ($)")
plt.xlabel("Category")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Key Insights & Recommendations**



* Top-performing category: Technology leads in both sales and profit.


* Underperforming sub-category: Tables have high sales but consistently low profit—possibly due to discounting or cost structure.


* Regional differences: The West region has the highest revenue, while the South lags behind.

* Monthly trend: Sales peak in November, suggesting holiday season performance boost.
