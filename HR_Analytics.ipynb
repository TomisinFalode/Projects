```python
# Data manipulation
import pandas as pd
import numpy as np
```

```python
# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set visualization style
sns.set(style='whitegrid')
```

```python
# Ignore warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')

# Preview the data
df.head()
```

```python
# Shape of the dataset
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# General info about data types and nulls
df.info()
```

```python
# Basic statistical summary of numeric columns
df.describe()
```

```python
# Check for missing values
print("Missing values per column:\n", df.isnull().sum())
```

```python
# Check for duplicates
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
# Count of employees by attrition status
sns.countplot(data=df, x='Attrition', palette='Set2')
plt.title('Employee Attrition Distribution')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

```python
# Age distribution
sns.histplot(data=df, x='Age', bins=20, kde=True, color='skyblue')
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.countplot(data=df, x='Gender', hue='Attrition', palette='Set1')
plt.title('Attrition by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(data=df, x='Department', hue='Attrition', palette='viridis')
plt.title('Attrition by Department')
plt.xticks(rotation=45)
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()
```
