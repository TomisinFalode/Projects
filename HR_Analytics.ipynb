```python
# Data manipulation
import pandas as pd
import numpy as np
```

```python
# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set visualization style
sns.set(style='whitegrid')
```

```python
# Ignore warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')

# Preview the data
df.head()
```

```python
# Shape of the dataset
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# General info about data types and nulls
df.info()
```

```python
# Basic statistical summary of numeric columns
df.describe()
```

```python
# Check for missing values
print("Missing values per column:\n", df.isnull().sum())
```

```python
# Check for duplicates
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
# Count of employees by attrition status
sns.countplot(data=df, x='Attrition', palette='Set2')
plt.title('Employee Attrition Distribution')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

```python
# Age distribution
sns.histplot(data=df, x='Age', bins=20, kde=True, color='skyblue')
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.countplot(data=df, x='Gender', hue='Attrition', palette='Set1')
plt.title('Attrition by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(data=df, x='Department', hue='Attrition', palette='viridis')
plt.title('Attrition by Department')
plt.xticks(rotation=45)
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='JobRole', hue='Attrition', palette='coolwarm')
plt.title('Attrition by Job Role')
plt.xticks(rotation=45)
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(data=df, x='OverTime', hue='Attrition', palette='Set2')
plt.title('Attrition by OverTime Status')
plt.xlabel('OverTime')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', palette='pastel')
plt.title('Monthly Income by Attrition Status')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.show()
```

```python
sns.histplot(data=df, x='YearsAtCompany', hue='Attrition', bins=15, kde=True, palette='husl')
plt.title('Years at Company vs Attrition')
plt.xlabel('Years at Company')
plt.ylabel('Frequency')
plt.show()
```

```python
# Compute correlation matrix for numeric columns
corr = df.corr(numeric_only=True)
```

```python
# Plot the heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt=".1f", linewidths=1.0)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

## Key Insights from the HR Attrition Analysis

1. **OverTime**: A significantly higher number of employees who work overtime tend to leave the company. This suggests burnout or lack of work-life balance.

2. **Monthly Income**: Employees with lower monthly income show a higher rate of attrition. Compensation may be a key factor in retention.

3. **Job Role**: Certain job roles like Sales Representatives and Laboratory Technicians have higher attrition compared to others.

4. **Years at Company**: Newer employees (0â€“3 years at the company) are more likely to leave, which suggests a need for better onboarding or early engagement programs.

5. **Department & Gender**: While there are variations, department and gender appear less strongly correlated with attrition in this dataset.

---

## Business Recommendations

- **Review Overtime Policies**: Consider reducing mandatory overtime or providing incentives for overtime-heavy roles.

- **Salary Benchmarking**: Reevaluate compensation structures, especially for roles with higher turnover.

- **Role-Specific Retention Plans**: Develop targeted retention initiatives for high-risk roles like Sales and Lab Techs.

- **Improve Onboarding**: Strengthen early-stage employee engagement and mentorship to reduce early attrition.
